# Seperate features and labels

```python
import numpy as np

def normalize_keypoints(X):
    """Normalizuje keypointy względem nadgarstka (punkt 0) i rozmiaru dłoni."""
    X_npy = X.to_numpy()
    X_normalized = X_npy.copy().reshape(-1, 21, 3)

    for i in range(len(X_normalized)):
        # Przesuń tak, żeby nadgarstek był w (0, 0, 0)
        wrist = X_normalized[i, 0].copy()
        X_normalized[i] -= wrist

        # Skaluj względem odległości nadgarstek-środkowy palec (punkt 9)
        scale = np.linalg.norm(X_normalized[i, 9])
        if scale > 0:
            X_normalized[i] /= scale

    return X_normalized.reshape(-1, 63)
```

```python
df = pd.concat([df_cont, df_nail, df_other, df_stop])
X = df.drop(columns=['label'])
X_normalized = normalize_keypoints(X)
y = df[['label']]
```

```python
from sklearn.model_selection import train_test_split
TEST_SPLIT = 0.2

X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=TEST_SPLIT, random_state=42, stratify=y)
```

#Creating model

```python
import torch
import torch.nn as nn
import torch.optim as optim
import os
```

```python
EPOCHS = 100
BATCH_SIZE = 32
LEARNING_RATE = 0.001
```

```python
class GestureMLP(nn.Module):
  def __init__(self, input_size=63, hidden_size=128, num_classes=4):
    super(GestureMLP, self).__init__()
    self.model = nn.Sequential(
        nn.Linear(input_size, hidden_size),
        nn.ReLU(),
        nn.Linear(hidden_size, hidden_size  // 2),
        nn.ReLU(),
        nn.Linear(hidden_size // 2, num_classes)
    )

  def forward(self, x):
    return self.model(x)
```

```python
from torch.utils.data import DataLoader, TensorDataset
CATEGORIES = ['continue_recording_sign', 'nail_biting_sign', 'other', 'stop_recording_sign']
INPUT_SIZE = 63  # 21 keypointów × 3 współrzędne (x, y, z)
HIDDEN_SIZE = 128
NUM_CLASSES = len(CATEGORIES)
MODEL_PATH = os.path.join('gesture_mlp.pth')

# --- TRENING ---
def train():


    # Konwersja na tensory
    X_train_t = torch.tensor(X_train, dtype=torch.float32)
    X_test_t = torch.tensor(X_test, dtype=torch.float32)
    y_train_t = torch.tensor(y_train.values.flatten(), dtype=torch.long)
    y_test_t = torch.tensor(y_test.values.flatten(), dtype=torch.long)

    # DataLoadery
    train_dataset = TensorDataset(X_train_t, y_train_t)
    test_dataset = TensorDataset(X_test_t, y_test_t)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE)

    # Model, loss, optimizer
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Device: {device}\n")

    model = GestureMLP().to(device)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

    # Trening
    best_acc = 0.0
    for epoch in range(EPOCHS):
        model.train()
        train_loss = 0.0

        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)

            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch)
            loss.backward()
            optimizer.step()

            train_loss += loss.item()

        # Walidacja
        model.eval()
        correct = 0
        total = 0

        with torch.no_grad():
            for X_batch, y_batch in test_loader:
                X_batch, y_batch = X_batch.to(device), y_batch.to(device)
                outputs = model(X_batch)
                _, predicted = torch.max(outputs, 1)
                total += y_batch.size(0)
                correct += (predicted == y_batch).sum().item()

        acc = 100 * correct / total

        if acc > best_acc:
            best_acc = acc
            torch.save({
                'model_state_dict': model.state_dict(),
                'categories': CATEGORIES,
                'input_size': INPUT_SIZE,
                'hidden_size': HIDDEN_SIZE,
                'num_classes': NUM_CLASSES
            }, MODEL_PATH)

        if (epoch + 1) % 10 == 0:
            print(f"Epoch [{epoch+1}/{EPOCHS}] Loss: {train_loss/len(train_loader):.4f} | Acc: {acc:.1f}% | Best: {best_acc:.1f}%")

    print(f"\nTrening zakończony!")
    print(f"Najlepsza dokładność: {best_acc:.1f}%")
    print(f"Model zapisany: {MODEL_PATH}")

```

```python
train()
```
